```python
# color code classes
# get class pred probabilities to use for plotting
def getColorMap(y_test, X_pred_probs):
    '''Returns color coded classes for plotting'''
    cmap = []
    alphas = []
    for y, probs_row in zip(y_test, X_pred_probs):
        if y == 0:
            cmap.append('b')
            alphas.append(probs_row[0])
        elif y == 1:
            cmap.append('r')     
            alphas.append(probs_row[1])
        else:
            cmap.append('g')     
            alphas.append(probs_row[2])
            
    return cmap, alphas
```

The plot shows that the principal components of each of the new york times categories point in spatially seperate directions, with some overlap near the origiin. The saturation of the colors directly correspond to probabilities, indicating how confident the model is about the predicted categories.

The 3 categories are very distinct. The is due to a couple of reasons. First, the content of a business article must indeed be considerably different than the content of a U.S or World article. Second, how we choose to vectorize the text data has a huge influence in how a machine learning model performs. We chose to vectorize the data using TfIdf. 

TfIdf stands for Term Frequency Inverse Document Frequency. TfIdf assigns importants of terms in a document in two ways. First if identifies the term frequencies in each document, term frequency tf(t,d). Where t stands for term, and d stands for document (in our case nyt article). So each term is given a frequency score for each document. Then the inverse document frequency idf(t,d) identifies how often a term occurs in other documents. However, idf gives lower scores for terms with high document frequency and high scores for terms with low document frequency. 

As a result, words like "the" will have a high *Tf* but will also likely appear in every document in the corpus, having a high *idf*. So the IDf for that word, in a corpus with n documents, will be zero. 

$$log\frac{n}{n} = 0$$
Idf does a log transformation on the inverse document frequency. This is done so that a term with a document frequency of 10 is ranked as more important than a term with document frequency of 2. However terms with document frequencies 100 and 105 will be weighted about the same; they are almost equally unimportant. 

A word like "Obama" will likely only be mentioned a few times in an article while also being mentioned in a subset of articles, but not the entire corpus. So the term "Obama" will be given a low Tf per article, but a high Idf. Thus the term "Obama" will be given a higher score indicating importance: this term contains more information than a word like "the".

```python
def get_tfidf_term_scores(feature_names):
    '''Returns dictionary with term names and total tfidf scores for all terms in corpus'''
    term_corpus_dict = {}
    # iterate through term index and term 
    for term_ind, term in enumerate(feature_names):
        term_name = feature_names[term_ind]
        term_corpus_dict[term_name] = np.sum(X_test_tfidf.T[term_ind].toarray())
        
    return term_corpus_dict
```

```python
# list of features created by tfidf
feature_names = tfidf.get_feature_names()
```

```python
term_corpus_dict = get_tfidf_term_scores(feature_names)
```

```python
def get_sorted_tfidf_scores(term_corpus_dict):
    '''Returns sort words from highest score to lowest score'''
    # sort indices from words wit highest score to lowest score
    sortedIndices = np.argsort( list(term_corpus_dict.values()))[::-1]
    # move words and score out of dicts and into arrays
    termNames = np.array(list(term_corpus_dict.keys()))
    scores = np.array(list(term_corpus_dict.values()))
    # sort words and scores
    termNames = termNames[sortedIndices]
    scores = scores[sortedIndices]
    
    return termNames, scores
```

```python
termNames, scores = get_sorted_tfidf_scores(term_corpus_dict)
```

```python
def getSelectScores(selectTerms):
    '''Returns a list of tfidf scores for select terms that are passed in'''
    score = [ term_corpus_dict[select_term]  for select_term in selectTerms]
    return score
```

```python
def drop_stopwords(corpus, stopWords):
    '''Return tokenized corpus with stopWords removed'''
    corpus_copy = []

    for doc in corpus:
        docs_copy = []
        for token in doc:
            token = token.lower()
            if token not in stopWords and token != 's':
                docs_copy.append(token)
        corpus_copy.append(docs_copy)
        
    return corpus_copy
```

```python
def buildSentanceVector(text, size):
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in text:
        try:
            vec += w2v[word].reshape((1, size))
            count += 1.
        except KeyError:
            continue
    if count != 0:
        vec /= count
    return vec
```

```python
def clean_data(data, stopwords):
    # initlize regex tokenizer
    reg_tokenizer = RegexpTokenizer(pattern=r'\w+')
    # tokenize corpus
    df = pd.DataFrame(data)
    token_data = df[0].replace(to_replace=r'[0-9]+', value="", regex=True)\
                    .apply(lambda doc: reg_tokenizer.tokenize(doc))
    # remove stopwords
    clean_data = drop_stopwords(token_data, stopwords)
    
    return clean_data
```

```python
from sklearn.decomposition import LatentDirichletAllocation
#from gensim.models import ldamodel
```

```python
# write class that returns the 5 data objects we need for the data viz
class topic_model_distributions(object):
    '''This class fits Latent Dirichlet Allocation for topic modeling
       and returns topic-term dist, doc-topic dist, and other distributions
       needed for the pyLDAvis visulization.'''
    
    def __init__(self, term_freq_data, stopwords,n_topics):
        
        self.lda_model = None
        self.term_freq_data = term_freq_data
        self.stopwords = stopwords
        self.doc_topic_dist = None
        self.doc_lengths = None
        self.vocab = None
        self.doc_topic_probs = []
        self.topic_indices = []
        self.topic_word_count = dict()
        self.topic_term_dists_dict = {}
        self.topic_term_dists_matrix = None
        self.smoothing_prob = 0.001
        self.lda_model = LatentDirichletAllocation(n_topics=n_topics, 
                                           learning_method="online", 
                                           n_jobs=3)        
    def fit_lda(self):
        self.lda_model.fit(self.term_freq_data)
        
    def set_doc_topic_dist(self):
        self.doc_topic_dist = self.lda_model.transform(self.term_freq_data)
        
        
        
    def set_topic_weight_and_index_(self):
        '''Sets probability that doc_i is a memeber of topic_j
           Sets topic index for each doc, i.e. doc_i is a member of topic_j'''

        for doc_probs in self.doc_topic_dist:
            
            # get topic index for each doc
            topic_index = np.argmax(doc_probs)
            self.topic_indices.append(topic_index)   
            
            # get probability that doc belongs to top topic
            doc_topic_prob = doc_probs[topic_index]
            self.doc_topic_probs.append(doc_topic_prob)


        
    def set_topic_word_count_(self):
        '''Sets total number of words that appear in each topic'''

        for doc_index, topic_index in enumerate(self.topic_indices):
            if topic_index not in self.topic_word_count.keys():
                self.topic_word_count[topic_index] = np.sum(self.term_freq_data[doc_index])
            else:
                self.topic_word_count[topic_index] += np.sum(self.term_freq_data[doc_index])
              
    def set_topic_term_matrix(self):
        '''Sets the topic-term distribution matrix'''
        
        self.set_topic_weight_and_index_()
        self.set_topic_word_count_()
        
        
        # iterate through topics ids for all docs in corpus
        for doc_index, topic_index in enumerate(self.topic_indices):
            # term frequency for doc i of topic j 
            doc_i_term_freq = self.term_freq_data[doc_index].toarray()[0]
            # normalized probabilities of words in doc i of topicj 
            # i.e. these words now have probs that correspond to their prob of appear in topic j
            doc_i_term_freq_normalized = doc_i_term_freq/self.topic_word_count[topic_index] 

            if topic_index not in self.topic_term_dists_dict.keys():
                self.topic_term_dists_dict[topic_index] = doc_i_term_freq_normalized
            else:
                self.topic_term_dists_dict[topic_index] += doc_i_term_freq_normalized

        self.topic_term_dists_matrix = np.vstack(self.topic_term_dists_dict.values())        

        
    def get_doc_topic_dist(self):
        '''Returns the doc topic probability distributions'''
        return  self.doc_topic_dist
    
    def get_topic_term_matrix(self):
        """Returns the topic-term probability distribution matrix"""
        return self.topic_term_dists_matrix
    
    def get_doc_lengths(self):
        '''Returns the number of words in each document'''
        # doc_lengths  -  the number of words in each document - shape `n_docs`
        return [doc.sum() for doc in self.term_freq_data]
        
```

```python
from sklearn.model_selection import ShuffleSplit
```

```python
N_topics = np.linspace(2,20, 19)
```

```python
def get_lda_perplexity(termFreq):
    
    ss = ShuffleSplit(n_splits=1, test_size=0.2, random_state=1234)
    
    for train_index, test_index in ss.split(termFreq):
        corpus_train = termFreq[train_index]
        corpus_test = termFreq[test_index]
    
    train_perplexity = []
    test_perplexity = []
    
    for n_topics in N_topics:
        lda = LatentDirichletAllocation(n_topics=n_topics, max_iter = 20, learning_method="online", n_jobs=3)
        print ("Fitting {} topics".format(n_topics))
        lda.fit(corpus_train);
        train_perplexity.append(lda.perplexity(corpus_train))
        test_perplexity.append(lda.perplexity(corpus_test))
        
    return train_perplexity, test_perplexity
```

```python
# train on tf*idf
start = time()
train_perplexity, test_perplexity = get_lda_perplexity(tfidf_UScorpus)
end = time()
```

```python
def get_entities(topic_ids, corpus):

    results = dict()
    unique_entities_dict = {}
    doc_ids = list(range(len(corpus)))
    
    for topic_id, doc_id, doc in zip(topic_ids, doc_ids, corpus):
        tokenized = nltk.word_tokenize(doc)
        tagged = nltk.pos_tag(tokenized)

        namedEnt_doc_i = nltk.ne_chunk(tagged, binary=False)

        for entity in namedEnt_doc_i:
            if isinstance(entity, nltk.tree.Tree):
                etext = " ".join([word for word, tag in entity.leaves()])
                label = entity.label()
            else:
                continue

            if label == 'PERSON':
                key = 'persons'
            elif label == 'ORGANIZATION':
                key = 'organizations'
            elif label == 'LOCATION':
                key = 'locations'
            elif label == 'GPE':
                key = 'other'
            else:
                key = None

            if topic_id not in results.keys():
                results[topic_id] = defaultdict(list)
                
            if topic_id not in unique_entities_dict.keys():
                unique_entities_dict[topic_id] = defaultdict(set)
                
             
            # store all types of named entitis
            # topic_i : {entity_type: entity}
            results[topic_id][key].append(etext.lower()) 
            
            # store only person entities
            if key == 'persons':
                # topic_i : {doc_j : (set of entities that appear in doc_j) } 
                unique_entities_dict[topic_id][doc_id].add(etext.lower())
                
            
    return results, unique_entities_dict
```

```python
def get_name_entity_count(entity_results):
    '''Returns the count of occurances of each name entity in each topic'''

    # { topic_id : { name_i: count_i  }  }
    persons_dict = dict()

    for topic_id, entity_dict in entity_results.items():
        persons_dict[topic_id] = Counter(entity_dict['persons'])

    return persons_dict
```

```python
# TODO -- REFACTOR CODE 

# get total num of entities in each topic

def get_entity_topic_count(entity_dict):
    '''Returns a dict where keys are topic ids and values are sum 
       of all entities that appear in the topic'''
    
    entity_count_dic = {}
    topic_ids = entity_dict.keys()

    for topic_id in topic_ids:
        entity_count_dic[topic_id] = len(entity_dict[topic_id])
        
    return entity_count_dic, topic_ids
```

```python
# clean named_entity_results
def flatten_results(entity_dict_data):
    '''Returns entity dict with flatten values. 
    
       The extract entities in the get_entities function returned nested lists
       in the dict values. Here we are correcting the format.'''
    
    
    # make deep copy of dict
    # so original dict is not over written
    entity_dict = copy.deepcopy(entity_dict_data)
    
    topics = entity_dict.keys()
    for topic in topics:
        flattend_values = np.array(list(entity_results[topic].values())).flatten()[0]
        
        entity_dict[topic] = flattend_values
        
    return entity_dict
```

```python
def invert_nested_dict(data):
    '''Returns the inverstion of a nested dictionary'''
    
    flipped = defaultdict(dict)
    for key, val in data.items():
        for subkey, subval in val.items():
            flipped[subkey][key] = subval
    return flipped
```

```python
def get_docs_with_co_occurances( co_occurance_entities_set, unique_entities_dict, topic_ids):
    '''Returns list of doc ids for co-occurances of entites'''
    
    co_occurances_doc_ids = []
    
    for topic_id in topic_ids:
        for doc_id, set_of_entities in unique_entities_dict[topic_id].items():
            
            if set_of_entities.issuperset(co_occurance_entities_set):
                co_occurances_doc_ids.append(doc_id)
                
    return co_occurances_doc_ids
```

```python
co_occurance_entities_set = set(('bernie sanders', 'paul ryan'))
co_occurances_doc_ids = get_docs_with_co_occurances(co_occurance_entities_set, unique_entities_dict, topic_ids)
```

```python
# TODO -- refactor code so doc_ids are given for each topic, for each co-occurance
def get_num_co_occurances(co_occurance_entities_set, unique_entities_dict, topic_ids, co_occurances_doc_ids):
    n_entities = 2
    
    co_entity_dict = defaultdict(dict)
    co_entity_doc_ids_list = []
    for co_entities in combinations(query_list, n_entities):
        co_entity_doc_ids = get_docs_with_co_occurances(set(co_entities), unique_entities_dict, topic_ids)
        
        co_entity_dict[co_entities[0]][co_entities[1]] = len(co_entity_doc_ids)
        co_entity_dict[co_entities[1]][co_entities[0]] = len(co_entity_doc_ids)
        
        co_entity_doc_ids_list.append([set(co_entities), co_entity_doc_ids])
        
    return co_entity_dict, co_entity_doc_ids_list
```
