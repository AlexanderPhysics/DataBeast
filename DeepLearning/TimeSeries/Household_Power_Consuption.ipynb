# Household Power Consuption

## Abstract

We perform KMeans clustering on four years of a single household's energy consumption in an attempt to discover load architype curves. Cluster metrics show that there is no indication of an optimium number of clusters. Using load architype curves, for this dataset, as the basis for future work will heavily rely on domain expertise. We then use a LSTM neural network for time-series prediction of energy consumption 24 hours into the futures. The results are impressive. 

## Introduction 

The work in this notebook is inspired by the data science worked performed at OPower. The Data Science team at OPower clustered over 800,000 daily household energy consumption curves and discovered that daily energy consupmtion curves fall into one of [five load architype curves](https://blog.opower.com/2014/10/load-curve-archetypes/).

The intention of this notebook is to expand on this work. Since we do not have access to OPower's colossal energy meter data set, we will use energy meter data from the [UCI Machine Learning repository](http://archive.ics.uci.edu/ml/datasets/Individual+household+electric+power+consumption).

## Goals

1) Create our own load architype curves for our data set.

2) Assign each day of energy consumpution to one of the architype curves

3) Perform time-series prediction of power consumption

```python
def normalize_data_types(df):
    cols = df.columns[2:]
    #Convert Series datatype to numeric, getting rid of any non-numeric values
    for col in cols:
        df[col] = pd.to_numeric(df[col], errors="coerce")   
    df.dropna(axis=0, inplace=True)
    return df
```

```python
# Get energy consumed not accounted for my submeters
def get_unaccounted_energy_load(df):
    Global_active_power = df["Global_active_power"]
    Global_reactive_power = df["Global_reactive_power"]
    sub_meter_one = df["Kitchen_WH"]
    sub_meter_two = df["LaundryRoom_WH"]
    sub_meter_three = df["HeatingCooling_WH"]
    unaccounted_energy = Global_active_power*1000./60. - sub_meter_one - sub_meter_two - sub_meter_three
    df["Unaccounted_energy_WH"] = unaccounted_energy
    df["Global_active_energy_WH"] = Global_active_power*1000./60.
    df["Global_reactive_power_WH"] = Global_reactive_power*1000./60.
    return df
```

```python
# create day vectors for clustering 
def get_smooth_data_by_day(data):
    energy_master = []
    energy_day = []
    energy_hour = []
    date_list = []
    
    for i, row in enumerate(data):
        energy_hour.append(row[2])
        
        # build day
        if i % 60 == 0 and i != 0:
            energy_day.append(np.mean(energy_hour))
            energy_hour = []
            
        # build master
        if i % 1440==0 and i!=0:
            energy_master.append(energy_day)
            date_list.append(row[0])
            energy_hour = []
            energy_day = []
            
    energy_master = np.array(energy_master)
    return energy_master, date_list
```

```python
# create day vectors for clustering 
def get_data_by_day(data):
    energy_list = []
    build_day = []
    date_list = []
    
    for i, row in enumerate(data):
        if i % 1440==0 and i!=0:
            # Sanity Check: every day will be unique
            #print row[0]
            energy_list.append(build_day)
            date_list.append(row[0])
            build_day = []
            
        build_day.append(row[2])
    energy_list = np.array(energy_list)
    
    return energy_list, date_list
```

```python
def get_cluster_data(k_clusters, day_vectors):
    k_results = []
    for k in k_clusters:
        km = KMeans(n_clusters=k, random_state=4)
        labels = km.fit_predict(day_vectors)

        # assign day vectors to their cluster label
        cluster_dict = defaultdict(list)
        for label, day_vector in zip(labels, day_vectors):
            cluster_dict[label].append(day_vector)

        # get average day vector for each cluster
        mean_days = np.array([np.mean(cluster_dict[key], axis=0) for key in cluster_dict.iterkeys()])
        k_results.append((k, mean_days))

    return k_results
```

```python
def cluster_metrics(estimator,  data):
    estimator.fit(data)
    sil_score =  metrics.silhouette_score(data, estimator.labels_,
                                      metric='euclidean')
    return estimator.inertia_, sil_score
```

```python
def get_cluster_std(labels, day_vectors):

    std_dict = defaultdict(list)
    cluster_dict = defaultdict(list)

    # assign day vectors to their cluster label
    for label, day_vector in zip(labels, day_vectors):
        cluster_dict[label].append(day_vector)

    for label in np.unique(labels):
        std_days = np.std(cluster_dict[label], axis=0)
        std_dict[label].append(std_days)

    return std_dict
```

```python
def get_upper_lower_std(load_curves,k):
    std_ub_dict = defaultdict(list)
    std_lb_dict = defaultdict(list)
    for k in range(0,k):
        std = std_dict[k]
        mean = load_curves[k]
        ub = mean + std
        lb = mean - std
        std_ub_dict[k] = ub[0]
        std_lb_dict[k] = lb[0]
    return std_ub_dict, std_lb_dict
```

```python
def dispersion (data, k):
    if k == 1:
        # get mean along the cols
        cluster_mean = np.mean(data, axis=0)
        # get sum of squared difference along the rows
        distances_from_mean = np.sum((data - cluster_mean)**2,axis=1)
        # get log of sum of distances from mean
        dispersion_val = np.log(sum(distances_from_mean))
    else:
        k_means_model_ = KMeans(n_clusters=k, max_iter=50, n_init=5, random_state=4).fit(data)
        # initialize list
        distances_from_mean = range(k)
        for i in range(k):
            distances_from_mean[i] = int()
            for idx, label in enumerate(k_means_model_.labels_):
                # if data row belongs to cluster 
                if i == label:
                    # get distance fromt point to centroid 
                    distances_from_mean[i] += sum((data[idx] - k_means_model_.cluster_centers_[i])**2)
        # get log of sum of distances from mean
        dispersion_val = np.log(sum(distances_from_mean))

    return dispersion_val

def reference_dispersion(data, num_clusters, num_reference_bootstraps):
    dispersions = [dispersion(generate_uniform_points(data), num_clusters) for i in range(num_reference_bootstraps)]
    mean_dispersion = np.mean(dispersions)
    stddev_dispersion = float(np.std(dispersions)) * np.sqrt(1. + 1. / num_reference_bootstraps) 
    return mean_dispersion, stddev_dispersion

def generate_uniform_points(data):
    mins = np.argmin(data, axis=0)
    maxs = np.argmax(data, axis=0)

    num_dimensions = data.shape[1]
    num_datapoints = data.shape[0]

    # initalize list 
    reference_data_set = np.zeros((num_datapoints,num_dimensions))
    for i in range(num_datapoints):
        for j in range(num_dimensions):
            reference_data_set[i][j] = random.uniform(data[mins[j]][j], data[maxs[j]][j])

    return reference_data_set   

def gap_statistic (data, nthCluster, referenceDatasets):
    # get dispersion for nthCluster
    actual_dispersion = dispersion(data, nthCluster)
    # get reference dispersion for nthCluster
    ref_dispersion, stddev_dispersion = reference_dispersion(data, nthCluster, num_reference_bootstraps)
    return actual_dispersion, ref_dispersion, stddev_dispersion
```

```python
class LSTM(object):

    def __init__(self, nin, n_hidden, nout):
        rng = np.random.RandomState(1234)
        # cell input
        W_ug = np.asarray(rng.normal(size=(nin, n_hidden), scale= .01, loc = 0.0), dtype = theano.config.floatX)
        W_hg = np.asarray(rng.normal(size=(n_hidden, n_hidden), scale=.01, loc = 0.0), dtype = theano.config.floatX)
        b_g = np.zeros((n_hidden,), dtype=theano.config.floatX)
        # input gate equation
        W_ui = np.asarray(rng.normal(size=(nin, n_hidden), scale =.01, loc=0.0), dtype = theano.config.floatX)
        W_hi = np.asarray(rng.normal(size=(n_hidden, n_hidden), scale =.01, loc=0.0), dtype = theano.config.floatX)
        b_i = np.zeros((n_hidden,), dtype=theano.config.floatX)
        # forget gate equations
        W_uf = np.asarray(rng.normal(size=(nin, n_hidden), scale =.01, loc=0.0), dtype = theano.config.floatX)
        W_hf = np.asarray(rng.normal(size=(n_hidden, n_hidden), scale =.01, loc=0.0), dtype = theano.config.floatX)
        b_f = np.zeros((n_hidden,), dtype=theano.config.floatX)
        # cell output gate equations
        W_uo = np.asarray(rng.normal(size=(nin, n_hidden), scale =.01, loc=0.0), dtype = theano.config.floatX)
        W_ho = np.asarray(rng.normal(size=(n_hidden, n_hidden), scale =.01, loc=0.0), dtype = theano.config.floatX)
        b_o = np.zeros((n_hidden,), dtype=theano.config.floatX)
        # output layer
        W_hy = np.asarray(rng.normal(size=(n_hidden, nout), scale =.01, loc=0.0), dtype = theano.config.floatX)
        b_hy = np.zeros((nout,), dtype=theano.config.floatX)

        # cell input
        W_ug = theano.shared(W_ug, 'W_ug')
        W_hg = theano.shared(W_hg, 'W_hg')
        b_g = theano.shared(b_g, 'b_g')
        # input gate equation
        W_ui = theano.shared(W_ui, 'W_ui')
        W_hi = theano.shared(W_hi, 'W_hi')
        b_i = theano.shared(b_i, 'b_i')
        # forget gate equations
        W_uf = theano.shared(W_uf, 'W_uf')
        W_hf = theano.shared(W_hf, 'W_hf')
        b_f = theano.shared(b_f, 'b_f')
        # cell output gate equations
        W_uo = theano.shared(W_uo, 'W_uo')
        W_ho = theano.shared(W_ho, 'W_ho')
        b_o = theano.shared(b_o, 'b_o')
        # output layer
        W_hy = theano.shared(W_hy, 'W_hy')
        b_hy = theano.shared(b_hy, 'b_hy')

        self.activ1 = T.nnet.sigmoid
        self.activ2 = T.tanh
        
        lr = T.scalar()
        u = T.matrix()
        t = T.scalar()


        h0_tm1 = theano.shared(np.zeros(n_hidden, dtype=theano.config.floatX))
        s0_tm1 = theano.shared(np.zeros(n_hidden, dtype=theano.config.floatX))

        
        
        
        

        #theano.printing.debugprint([h0_tm1, u, W_hh, W_uh, W_hy, b_hh, b_hy], print_type=True)
        [h, s], _ = theano.scan(self.recurrent_fn, sequences = u,
                           outputs_info = [h0_tm1, s0_tm1],
                           non_sequences = [W_ug, W_hg, b_g, W_ui, W_hi,
                                            b_i, W_uf, W_hf, b_f, W_uo, W_ho, b_o, W_hy, b_hy])

        y = T.dot(h[-1], W_hy) + b_hy
        cost = ((t - y)**2).mean(axis=0).sum()
        update_params = [W_ug, W_hg, b_g, W_ui, W_hi, b_i,W_uf, W_hf, b_f, W_uo, W_ho, b_o, W_hy, b_hy]

        gW_ug, gW_hg, gb_g, gW_ui, gW_hi, gb_i,gW_uf, gW_hf, gb_f, gW_uo, gW_ho, gb_o, gW_hy, gb_hy\
        = T.grad(cost,update_params)
            

            
                 
            
        update = [(W_ug, W_ug - lr*gW_ug), 
                  (W_hg, W_hg - lr*gW_hg ), 
                  (b_g, b_g - lr*gb_g), 
                  (W_ui, W_ui - lr*gW_ui),
                  (W_hi, W_hi - lr*gW_hi), 
                  (b_i, b_i - lr*gb_i), 
                  (W_uf, W_uf - lr*gW_uf), 
                  (W_hf, W_hf - lr*gW_hf),
                  (b_f, b_f - lr*gb_f),
                  (W_uo, W_uo - lr*gW_uo), 
                  (W_ho, W_ho - lr*gW_ho), 
                  (b_o, b_o - lr*gb_o),
                  (W_hy, W_hy - lr*gW_hy), 
                  (b_hy, b_hy - lr*gb_hy)]
        
        #theano.printing.debugprint([h0_tm1], print_type=True)
        self.train_step = theano.function([u, t, lr], cost,
            on_unused_input='warn',
            updates=update,
            allow_input_downcast=True)
        
        
                
        self.predict_step = theano.function([u, t], outputs=[cost, y],
           on_unused_input='warn',
           allow_input_downcast=True)
        

    def recurrent_fn(self, u_t, h_tm1, s_tm1, W_ug, W_hg, b_g, W_ui, W_hi,
                                            b_i, W_uf, W_hf, b_f, W_uo, W_ho, b_o, W_hy, b_hy):
        g_t = self.activ2(T.dot(u_t, W_ug) + T.dot(h_tm1, W_hg) + b_g)
        i_t = self.activ1(T.dot(u_t, W_ui) + T.dot(h_tm1, W_hi) + b_i)
        f_t = self.activ1(T.dot(u_t, W_uf) + T.dot(h_tm1, W_hf) + b_f)
        o_t = self.activ1(T.dot(u_t, W_uo) + T.dot(h_tm1, W_ho) + b_o)
        s_t = g_t * i_t + s_tm1*f_t
        h_t = self.activ2(s_t)*o_t
        
        #h_t = self.activ2(T.dot(h_tm1, W_hh) + T.dot(u_t, W_uh) + b_hh)
        return [h_t, s_t]
    
```

$\textbf{Inertia}$ is a measure of the within-cluster sum of squares, $W_{k}$. This sums up the distance between from each point in the cluster and the centriod. As expected, the inertia decreases monotonically with increasing number of clusters. With more clusters present, each point is more likely to be closer to a centroid. The Inertial vs. Clusters plot is often used to indentify the optimizal number of clusters by looking for a kink or 'elbow' on the curve. Unfortunately, this curve is smooth, has no kink, and so proves to be inconclusive. 

$\textbf{Sihouette Score}$ is a measure of how similar each point is to all other points within the same cluster, as well as how dissimilar each point is to points in other clusters. A Sihouette score of 1 is ideal, a score of -1 is the worest case, and a score of 0 indicate overlapping clusters. A weakness of the score is when the true data has no discernible clusters or only one true cluster. The Sihouette Score vs. Clusters plot shows that all clusters have a score near to 0, indicating that there isn't a sharp and distinct difference between the clusters. Unfortunately, this curve also proves to be inconclusive. 

$\textbf{Gap Statistic}$ uses a reference null distribution of the data (random sampling) and compares the dispersion on our data vs the reference. The ideal value of k will then be the value for which  $log(W_{k})$  drops furthest below this reference curve $log(W_{kb})$. The first instance in which the following condition is met, indicates the optimum number of clusters, $\mathrm{Gap}(k) \geq \mathrm{Gap}(k+1) - s_{k+1}$. The Log Interia vs. Clusters plot shows actual dispersion of the clusters $log(W_{k})$ and the reference dispersion $log(W_{kb})$. The Gap Statistic plot shows the difference between those values. Yet again, the Gap Statistic proves to be inconclusive for a confident answer for the optimum number of clusters.

## Time-Series Prediction

Here we will prep the data for time-series prediction. Unlike none time-sereis data, we cannot randomly shuffle the data and use KFold cross validation. In order to preserve the time dependence, we will split the data into a train and test set where the test set will be the last 20% of the data. 

As mentioned earlier, we'll be using the noisy minute-by-minute day vectors. The noise helps neural nets train better and provides a larger data set than the smooth day vectors, which neural nets need to train well.  

```python
def create_dataset(dataset, look_back=1, look_ahead=1440):
	dataX, dataY = [], []
	for i in range(len(dataset)-look_ahead):
		a = dataset[i:(i+look_back), 0]
		dataX.append(a)
		dataY.append(dataset[i + look_ahead, 0])
	return np.array(dataX, dtype = theano.config.floatX), np.array(dataY, dtype = theano.config.floatX)
```

```python
day_split = 1150
# flatten day vectors 
kW_energy_train = day_vectors[:day_split].flatten()
kW_energy_test = day_vectors[day_split:].flatten()

kW_energy_train = kW_energy_train.reshape(kW_energy_train.shape[0], 1)
kW_energy_test = kW_energy_test.reshape(kW_energy_test.shape[0], 1)
```

```python
# format train and test for 24hour/1440minute predictions
X_train, Y_train  = create_dataset(kW_energy_train,look_back=1, look_ahead=1440)
X_test, Y_test  = create_dataset(kW_energy_test, look_back=1, look_ahead=1440)
```

```python
lr = 0.01 # learning rage for gradient descent
train_mse = []
test_mse = []
train_length = len(X_train) 
test_length = len(X_test) 
```

```python
# initalize LSTM model
nInputs = X_train.shape[1]
n_hidden = 20
nout = 1
lstm = LSTM( nInputs, n_hidden, nout)
```
